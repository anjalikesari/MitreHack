```python
# Importing Necessary Data and Initial Observation
import pandas as pd
import numpy as np
import re
!pip install tldextract
import tldextract

data = pd.read_csv('/content/drive/MyDrive/data/Phishing_Mitre_Dataset_Summer_of_AI.csv')
test_data = pd.read_csv('/content/drive/MyDrive/data/Phishing_Mitre_Dataset_Summer_of_AI.csv')

# data = pd.read_csv('data/Phishing_Mitre_Dataset_Summer_of_AI.csv')
data.head(15)
```

```python
# Increasing readability of data
data = data.rename(columns={'create_age(months)': 'Domain Age (months)', 'expiry_age(months)': '# Months till Expiration (months)', 'update_age(days)': 'Last Time Domain was Updated (days)', 'Label': 'Phishing?'})
data
```

```python
# Replacing -1 with NaN Values (Mean of current column)
data = data.replace(-1, np.nan)
data = data.fillna(data.mean(axis=0));
data
```

```python
# Sorting values based on whether URL is Phishing or not
data = data.sort_values(by='Phishing?')
data['Phishing?'] = pd.to_numeric(data['Phishing?'])
data = data.reset_index(drop=True);
data
```

```python
# Length of Phishing websites
print("# Phishing Websites: ", len(data[data['Phishing?'] == 1]))
```

```python
# Examining the links of Phishing sites
data[data['Phishing?'] == 1]
data['URL']
```

```python
# Creating character counts for URL in new column
data["URL Length"] = data["URL"].str.len()
```

```python
data
```

```python
# Counting the number of digits in the url using the python library re
def num_digits(string):
  count = len(re.findall(r'\d', string))
  return count

data['Digits_Count'] = data['URL'].apply(lambda x: num_digits(x))

data
```

```python
# Does it have an @ symbol, yes - 1, no - 0
def find_at_symbol(string):
  return_value = string.find('@')
  if (return_value == -1):
    return 0
  else:
    return 1

data['@_symbol'] = data['URL'].apply(lambda x: find_at_symbol(x))

data
```

```python
# Checking to see if the url has a subdomain using package tldextract
# Note: don't know how we would want to use this, can alter this column, but for right now, just has the subdomain of each url
def find_subdomain(string):
  result = tldextract.extract(string)
  return result.subdomain

data['Subdomain'] = data['URL'].apply(lambda x: find_subdomain(x))

data
```

```python
# Check to see if an IP address is in the URL
def has_IP(string):
  ip_pattern = re.compile(r'\b(?:\d{1,3}\.){3}\d{1,3}\b')
  result = re.search(ip_pattern, string)
  return bool(result)

data['IP_Address'] = data['URL'].apply(lambda x: has_IP(x))

data
```

```python
# Check to see how many subdomains there are
def count_subdomain(string):
  if string:
    result = string
    if has_IP(string):
      ip_pattern = re.compile(r'\b(?:\d{1,3}\.){3}\d{1,3}\b')
      result = re.sub(ip_pattern, '', string)
    subdomains = result.split('.')
    return len(subdomains)
  else:
    return 0

data['Num_Subdomains'] = data['Subdomain'].apply(lambda x: count_subdomain(x))

data
```

```python
# Checking if top level domain (.com, .org) is a valid one
# If use this...probably need to do more research on the most common TLDs

VALID_TLD = ['com', 'org', 'net', 'int', 'edu', 'gov', 'mil', 'uk']

def valid_tld(string):
  result = tldextract.extract(string)
  tld = result.suffix
  return tld in VALID_TLD

data['valid_tld'] = data['URL'].apply(lambda x: valid_tld(x))

data
```

```python

models_dict = {'Decision_Tree': DecisionTreeClassifier(),
    'random_forest':     RandomForestClassifier(n_estimators=100, max_depth=15, max_features=3, bootstrap=True),
               'knn': KNeighborsClassifier(n_neighbors=7, metric='manhattan')}
```

```python
def evaluate_model(model, X, y):
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=2, random_state=1)
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, verbose=1, n_jobs=3, error_score='raise')
    return scores
model_scores = defaultdict()
for name, model in models_dict.items():
    print('Evaluating {}'.format(name))
    scores = evaluate_model(model, X, y)
    model_scores[name] = scores
```

```python
def plot_results(model_scores, name):

    model_names = list(model_scores.keys())
    results = [model_scores[model] for model in model_names]
    fig = go.Figure()
    for model, result in zip(model_names, results):
        fig.add_trace(go.Box(
            y=result,
            name=model,
            boxpoints='all',
            jitter=0.5,
            whiskerwidth=0.2,
            marker_size=2,
            line_width=1)
        )

    fig.update_layout(
    title='Performance of Different Models Using 5-Fold Cross-Validation (10 Trials)',
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
    xaxis_title='Model',
    yaxis_title='Accuracy',
    showlegend=False)
    fig.show()

plot_results(model_scores, name='base_models_cv')
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegressionCV
base_models = [('random_forest', RandomForestClassifier(n_estimators=300, max_depth=15, max_features=3, bootstrap=True)),
               ('SVC', SVC(C=1000, gamma="scale", kernel="rbf")),
               ('knn', KNeighborsClassifier(n_neighbors=7, metric='manhattan'))]
meta_model = RandomForestClassifier( max_depth=5)
stacking_model = StackingClassifier(estimators=base_models,
                                    final_estimator=meta_model,
                                    passthrough=True,
                                    cv=5,
                                    verbose=2)
```

```python
stacking_scores = evaluate_model(stacking_model, X, y)
model_scores['stacking'] = stacking_scores
```

```python
plot_results(model_scores, name='stacking_model_cv')
```
